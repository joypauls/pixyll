```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams["figure.figsize"] = (10,5)
```

```python
def exponential_function(a, start, end, n):
    t = np.linspace(start, end, n)
    return pd.DataFrame({"t": t, "base": "a = " + str(a), "f": a**t})
```

```python
for i in [0.05, .5, 1, 2, 3]:
    df = exponential_function(i, -0.3, 2, 100)
    plt.plot(df.t, df.f, label = "a=" + str(i), linewidth = 2)

plt.legend()
plt.vlines(0, 0, 9, linestyles="dashed", alpha=0.4)
plt.xlabel("t")
plt.ylabel("f(t)")
plt.show()
```
